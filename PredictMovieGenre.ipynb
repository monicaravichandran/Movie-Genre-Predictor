```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
import time

scifi_link = "https://www.imdb.com/list/ls056870923/"

war_link = "https://www.imdb.com/list/ls009678583/"

western_link = "https://www.imdb.com/list/ls002124326/"

action_link = "https://www.imdb.com/list/ls009668579/"

adventure_link = "https://www.imdb.com/list/ls009609925/"

```

```python
resp = requests.get(scifi_link)
soup = BeautifulSoup(resp.content, "html.parser")

# gets the movie name and the link of the image 
name_n_img_scifi = soup.find_all("img", {"class":"loadlate"})

rows = []
for movie in name_n_img_scifi:
    movie_name = movie.get("alt")
    movie_image_link = movie.get("loadlate")
    
    rows.append({
        "Movie Name": movie_name,
        "Poster Image Link": movie_image_link,
    })

scifi_df = pd.DataFrame(rows)
scifi_df = scifi_df[:100]
#scifi_df # contains movie name and the link of the poster of scifi movies
```

```python
movie_synopsis = soup.find_all("p", {"class":""})[2:]

rows = []
for movie in movie_synopsis:
    movie_description = movie.text
    #print(movie_description)
    
    rows.append({
        "Synopsis": movie_description.replace ("\n", ""),
    })

    
scifi_df_1 = pd.DataFrame(rows)
scifi_df_1 = scifi_df_1[:100]
#scifi_df_1 # contains the synopsis of scifi movies
```

```python
scifi_df["Synopsis"] = scifi_df_1["Synopsis"]
#scifi_df
```
